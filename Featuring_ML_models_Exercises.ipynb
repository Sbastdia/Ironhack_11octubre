### Exercise 2

 - Import fifa.csv . It contains data about football players
 - As you can see from columns [-28] onwards all the variables have a '+', use RegEx functions to clean these variables in the least number of steps as possible. Define a function if you find necessary
 - Variables Height and Weight are in feet and in lbs. Translate them into cm. Probably you will need some RegEx expression

```python
fifa = pd.read_csv('/content/drive/MyDrive/Ironhack/11octubre/fifa.csv')
fifa.head(3)
```

```python
import re

def limpiar_variables(dataframe):
    # Define una expresión regular para buscar un número seguido de un '+'
    pattern = r'(\d+)\+\d+'

    # Recorremos las columnas desde [-28] hacia delante
    for column in dataframe.columns[-28:]:
        # Eliminamos el '+' y el número siguiente
        dataframe[column] = dataframe[column].replace(pattern, r'\1', regex=True)

    return dataframe
```

```python
#Ahora que tenemos definida la función, limpiamos el dataframe
limpiar_variables(fifa)
fifa.head(3)
```

### HEIGHT

```python
# Utilizar expresiones regulares para extraer los números de pies y pulgadas
height_values = fifa['Height'].str.extract(r'(\d+)\'(\d+)["]')
height_values.columns = ['Feet', 'Inches']

# Rellenar los valores faltantes con ceros
height_values = height_values.fillna(0)

# Convertir las columnas "Feet" y "Inches" a números
height_values['Feet'] = height_values['Feet'].astype(int)
height_values['Inches'] = height_values['Inches'].astype(int)

# Realizar la conversión a centímetros
fifa['Height_in_cm'] = height_values['Feet'] * 30.48 + height_values['Inches'] * 2.54

# Eliminar la columna original "Height"
fifa.drop(columns=['Height'], inplace=True)

fifa.head(3)
```

### WEIGHT

```python
# Eliminar 'lbs' de la columna Weight y convertir a enteros
fifa['Weight'] = fifa['Weight'].str.replace('lbs', '').astype(int)

# Realizar la conversión de libras a kilogramos (1 libra = 0.453592 kilogramos)
fifa['Weight_in_kg'] = fifa['Weight'] * 0.453592

# Eliminar la columna original "Height"
fifa.drop(columns=['Weight'], inplace=True)

fifa.head(3)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RepeatedStratifiedKFold, cross_val_score, GridSearchCV, RepeatedKFold
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, AdaBoostClassifier, AdaBoostRegressor
```

```python
possum.columns
```

```python
scaler = StandardScaler()

possum=scaler.fit_transform(possum)
```

```python
possum= pd.DataFrame(possum, columns = ['case', 'site', 'age', 'hdlngth', 'skullw', 'totlngth', 'taill',
       'footlgth', 'earconch', 'eye', 'chest', 'belly', 'Pop_Vic', 'Pop_other',
       'sex_f', 'sex_m'])
```

```python
possum.isna().sum()
```

```python
mean_value_2=possum['age'].mean()
possum['age'].fillna(value=mean_value_2, inplace=True)

mean_value_7=possum['footlgth'].mean()
possum['footlgth'].fillna(value=mean_value_7, inplace=True)
```

```python
X=possum.drop('skullw', axis=1)
y=possum['skullw']
X.head()
```

```python
X_train, x_test, y_train, y_test= train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
classifiers = {
    'AdaBoostRegressor': (AdaBoostRegressor(), {'n_estimators': [100]}), #[['case', 'site'], ['case', 'site', 'totlngth'],
    'RandomForestRegressor': (RandomForestRegressor(), {'max_depth': [None], 'criterion': ['absolute_error']})
}
```

```python
cross_val = RepeatedKFold(n_splits=5, random_state=1)
results = {}
for clf_name, (clf, params) in classifiers.items():
    grid_search = GridSearchCV(clf, params, cv=cross_val, scoring = 'neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    results[clf_name] = {
        'Best Parameters': grid_search.best_params_,
        'Best Score': grid_search.best_score_
    }
```

```python
for clf_name, result in results.items():
    print(f'Classifier: {clf_name}')
    print(f'Best Parameters: {result["Best Parameters"]}')
    print(f'Best Score: {result["Best Score"]}\n')
```
